```python
#question C
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
ob=pd.read_csv('data.csv')
ab_ord=ob.sort_values(by='Happiness Score',ascending =False)
ab_ord=ab_ord.head(10)
ab=ab_ord['Happiness Score']
ab1=ab_ord['Country']
plt.bar(ab,ab1)
plt.xlabel('Score')
plt.ylabel('Country')
plt.title('Happiness Score by Country')
plt.show()
```

```python
#Question C
plt.legend(loc='upper right')
plt.bar(ab,ab_ord['Economy'],color='silver',edgecolor='black',label='Economy')
plt.bar(ab,ab_ord['Family'],color='gold',edgecolor='black',bottom=ab_ord['Economy'],label='Family')
plt.bar(ab,ab_ord['Health'],color='green',edgecolor='black',bottom=ab_ord['Economy']+ab_ord['Family'],label='Health')
plt.bar(ab,ab_ord['Freedom'],color='red',edgecolor='black',bottom=ab_ord['Economy']+ab_ord['Family']+ab_ord['Health'],label='Freedom')
plt.bar(ab,ab_ord['Generosity'],color='blue',edgecolor='black',bottom=ab_ord['Economy']+ab_ord['Family']+ab_ord['Health']+ab_ord['Freedom'],label='Generosity')
plt.bar(ab,ab_ord['Corruption'],color='purple',edgecolor='black',bottom=ab_ord['Economy']+ab_ord['Family']+ab_ord['Health']+ab_ord['Freedom']+ab_ord['Generosity'],label='Corruption')
plt.bar(ab,ab_ord['Dystopia'],color='brown',edgecolor='black',bottom=ab_ord['Economy']+ab_ord['Family']+ab_ord['Health']+ab_ord['Freedom']+ab_ord['Generosity']+ab_ord['Corruption'],label='Dystopia')
```
